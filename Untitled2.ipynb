```python
import requests

patent_number = "US20240194707A1"  # US patent
url = f"https://api.patentsview.org/patents/query?q={{\"_eq\":{{\"patent_number\":\"{patent_number}\"}}}}&f=[\"patent_number\",\"patent_title\",\"patent_abstract\"]"

response = requests.get(url)
data = response.json()
patent = data["patents"][0]
print({
    "title": patent.get("patent_title", ""),
    "abstract": patent.get("patent_abstract", "")
})
```
